```python
import pandas as pd
```

```python
file_path = 'all_data.csv'
```

```python
GDP_data = pd.read_csv(file_path)
df = pd.DataFrame(GDP_data)
```

```python
df.head(20)
```

```python
Unique_Countries = df.Country.unique()
print(Unique_Countries)
```

```python
import matplotlib.pyplot as plt
column_headers = df.columns
```

```python
Countries_Chile = df[(df['Country'] == 'Chile') & (df['Year'] >= 2000) & (df['Year'] <= 2015)]
display(Countries_Chile)
```

```python
display(Countries_Chile)
```

```python
#Chart for Life Expectancy by Birth for Chile
plt.figure(figsize=(8, 6))  # Optional: Set the figure size
plt.plot(Countries_Chile['Year'] , Countries_Chile['Life expectancy at birth (years)'], color='blue', marker='o')
plt.title('Life Expectancy Projections for Chile')
plt.xlabel('Year')
plt.ylabel('Life expectancy at birth (years)')
plt.show()
```

```python
#Chart for Life Expectancy by Birth for China
Countries_China= df[(df['Country'] == 'China') & (df['Year'] >= 2000) & (df['Year'] <= 2015)]
```

```python
#Life Exprectanncy per country by Year
Countries = df[(df['Year'] >= 2000) & (df['Year'] <= 2015)]
display(Countries)
```

```python
plt.figure(figsize=(8, 6))  # Optional: Set the figure size
plt.scatter(Countries["Country"][Countries["Year"] == 2000] , Countries['Life expectancy at birth (years)'][Countries["Year"] == 2000], color='green', marker='o')
plt.title('Life Expectancy Projections for the year 2000')
plt.xlabel('Country')
plt.ylabel('Life expectancy at birth (years)')
plt.show()
```

```python
#Average Life Expectancy Variable
Avg_LE = df.groupby('Country')['Life expectancy at birth (years)'].mean()
display(Avg_LE)
```

```python
#Average Life Expectancy by Country
plt.figure(figsize=(8, 6))  # Optional: Set the figure size
plt.scatter(Avg_LE.index , Avg_LE , color='green', marker='o')
plt.title('Life Expectancy by Country')
plt.xlabel('Country')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
#GDP trend in countries
plt.figure(figsize=(8, 6))  # Optional: Set the figure size
plt.scatter(Countries["Country"] , Countries['GDP'], color='green', marker='o')
plt.title('GDP Trend')
plt.xlabel('Country')
plt.ylabel('GDP')
plt.show()
```
